```python
#將空值刪除
new_data = data.dropna()
```

```python
#切分input和output
#x作為input
#y作為output
#特徵欄位
x=new_data.loc[:,['age','sex','cp','trestbps','chol','fbs','restecg','thalach','exang','oldpeak','slope','ca','thal']]
#目標欄位
y=new_data.loc[:,['target']]
Feature = x;
Label = y;
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
feature_scaler = scaler.fit_transform(Feature)
```

```python
#切分資料集與測試集，test_size=0.33，random_state=1 
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(feature_scaler, Label, test_size = 0.33, random_state = 1)
```

```python
#使用sklearn.svm裡的SVC進行分析，kernel 設為'linear'，印出模型最終的準確度
from sklearn.svm import SVC
classifier = SVC(kernel='linear')
classifier.fit(X_test,Y_test.values.ravel())
print(classifier.score(X_test,Y_test.values.ravel()))
```
